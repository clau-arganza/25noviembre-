```python
print(2*(2**4))
print((2**2)**2)
print(2*(2**2))
print(2**2)
print(2**1)

def pow_fast(a,n):
    if n==0:
        return 1
    
    half=pow_fast(a,n//2)

    if n%2==0:
        return half*half
    else:
        return half*half*a
    
#ejemplo de uso

base=2
exponent=5
print(pow_fast(base,exponent))     #output:32

```
