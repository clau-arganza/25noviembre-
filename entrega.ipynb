```python
import math

# DISTANCIA ENTRE DOS PUNTOS

def dist(p1, p2):
    return math.dist(p1, p2)

# ALGORITMO DIVIDE Y VENCERÁS PARA EL PAR MÁS CERCANO

def closest_pair(points):
    px = sorted(points)                   # ordenar por x
    py = sorted(points, key=lambda p: p[1])  # ordenar por y
    return closest_rec(px, py)

def closest_rec(px, py):
    n = len(px)

    # Caso base -> usar fuerza bruta (3 puntos máximo)
    if n <= 3:
        return brute(px)

    mid = n // 2
    mid_point = px[mid]

    # Dividir en izquierda (Q) y derecha (R)
    Qx = px[:mid]
    Rx = px[mid:]
    Qy = [p for p in py if p[0] <= mid_point[0]]
    Ry = [p for p in py if p[0] > mid_point[0]]

    # Recursión
    d_left, pair_left = closest_rec(Qx, Qy)
    d_right, pair_right = closest_rec(Rx, Ry)

    # Mejor distancia entre ambas mitades
    if d_left < d_right:
        d = d_left
        best_pair = pair_left
    else:
        d = d_right
        best_pair = pair_right

    # Construir franja vertical
    strip = [p for p in py if abs(p[0] - mid_point[0]) < d]

    # Comprobar franja (solo 7 vecinos)
    for i in range(len(strip)):
        for j in range(i+1, min(i+7, len(strip))):
            dd = dist(strip[i], strip[j])
            if dd < d:
                d = dd
                best_pair = (strip[i], strip[j])

    return d, best_pair

# FUERZA BRUTA (solo para base recursiva)

def brute(points):
    min_d = float("inf")
    best_pair = None
    for i in range(len(points)):
        for j in range(i+1, len(points)):
            d = dist(points[i], points[j])
            if d < min_d:
                min_d = d
                best_pair = (points[i], points[j])
    return min_d, best_pair


# LISTA DE PUNTOS DEL EJEMPLO
points = [
(2,3),(5,4),(9,6),(4,7),(8,1),(7,2),(12,5),(14,3),(3,10),(6,8),
(11,11),(13,9),(15,6),(18,2),(17,7),(19,4),(21,3),(20,8),(23,5),
(25,7),(24,1),(26,4),(30,2),(28,9),(31,7),(34,6),(33,3),(36,8),(38,5)
]

# EJECUCIÓN
dist_min, pair = closest_pair(points)

print("Distancia mínima:", dist_min)
print("Par más cercano:", pair)
```
